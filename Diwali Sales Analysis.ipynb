EXPLORATORY DATA ANALYSIS

GENDER

```python
ax= sns.countplot(x='Gender',data=df,palette='viridis',hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Graph based on the total Amount spend by each gender on buying products

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_spnd = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y="Amount",data=sales_spnd,palette='dark:yellow')
plt.show
```

From above graph we can see that most od the buyers are Females and even the purchasing power of Females are greater than Men

AGE

```python
b = sns.countplot(x='Age_Group',data=df,hue="Gender")
for bars in b.containers:
    b.bar_label(bars)
plt.show
```

```python
# Total Amount vs Age Group
df.groupby(['Age_Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age = df.groupby(['Age_Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age_Group',y='Amount',data=sales_age,palette='viridis')
plt.show()
```

From above graphs we can see the most of the buyers are of age group between 26-35 years female

State

```python
# TOTAL NUMBER OF ORDERS FROM TOP 10 STATES
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
orders_st = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Orders',data=orders_st,palette='Accent')
plt.show
```

```python
# TOTAL AMOUNT/SALES FROM TOP 10 STATES
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Amount',data=sales_state,palette='Accent')
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharastra, and Karnatka respectively.

MARITAL STATUS

```python
ax = sns.countplot(x='Marital_Status',data=df,palette='Reds')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# TOTAL AMOUNT OF SALES BASED ON MARITAL STATUS
sales_marital = df.groupby(["Marital_Status",'Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y="Amount",data=sales_marital,hue='Gender',palette="viridis")
```

*From above graphs we can see that most of the buyers are married (womens) and they have high purchasing power

Occupation

```python
ax = sns.countplot(x='Occupation',data=df,palette='viridis')
sns.set(rc={'figure.figsize':(20,6)})
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
# TOTAL SALES vs OCCUPATION
sales_occupation = df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending=False)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='Occupation',y='Amount',data=sales_occupation,palette='viridis')
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT Sector, Healthcare, and Aviation field.

PRODUCT CATEGORY

```python
# TOTAL NUMBER OF ORDERS BY EACH PRODUCT CATEGORY
ax = sns.countplot(x='Product_Category',data=df,palette='viridis')
sns.set(rc={'figure.figsize':(34,6)})
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
# TOTAL SALES vs Product Category
sales_occupation = df.groupby(['Product_Category'],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='Product_Category',y='Amount',data=sales_occupation,palette='viridis')
plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing & Apparel, Electronics & Gadgets Category

PRODUCTS

```python
# TOP 10 MOST SOLD PRODUCTS
sold_prdts = df.groupby(['Product_ID'],as_index=False)["Orders"].sum().sort_values(by=["Orders"],ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,6)})
sold_prdts = sns.barplot(x='Product_ID',y='Orders',data=sold_prdts,palette='viridis')
for i in sold_prdts.containers:
    sold_prdts.bar_label(i)
plt.show()
```

CONCLUSION:

Married women having age group between 26-35 years from Uttar Pradesh, Maharastra, and Karnataka working in IT, Healthcare and Aviation Sector are more likely to buy products from Food, Clothing and Electronics Categories.
