EXPLORATORY DATA ANALYSIS

```python
df.columns
```

SIZE

```python
ax = sns.countplot(x='Size',data=df,palette='viridis')
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

*From above graph you can see most of the people buys M-Size

```python
s_qty = df.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by=['Quantity'],ascending=False)
sns.barplot(x='Size',y='Quantity',data=s_qty,palette='viridis')
plt.show()
```

*From above graph you can see that most of the buyed Quantity are M-Size

COURIER STATUS

```python
ax = sns.countplot(x='Courier Status',data=df,hue='Status',palette='viridis')
sns.set(rc={'figure.figsize':(25,5)})
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

*From above graph you can see that the majority of the orders are shipped through the courier.

```python
sns.displot(df['Size'])
plt.show()
```

CATEGORY

```python
# sns.displot(df['Category'],bins=10)
plt.hist(df['Category'],bins=10,color='r',edgecolor='Black')
plt.xticks(rotation=90)
plt.figure(figsize=(10,5))
plt.show()
```

*From above graph you can see that most of the buyers bought T-shirts.

B2B

```python
# Checking B2B Data by using Pie Chart
B2B = df['B2B'].value_counts()

plt.pie(B2B,labels=B2B.index,autopct='%1.1f%%')
plt.show()
```

*From above graph we can see that the maximum i.e. 99.2% of buyers are retailers and 0.8% are B2B buyers.

CATEGORY vs SIZE 

```python
# Checking for which cateogory what all sizes are available using scatter plot

sns.scatterplot(x='Category',y='Size',data=df)
sns.set(rc={'figure.figsize':(20,5)},style=("dark"))
plt.title("Scatter Plot",size=20,color='r')
plt.show()
```

*From above graph we can see what all sizes are available for each category like T-shirt is available in all sizes except Free Size.

STATES AND CITIES

```python
# TOP 10 STATES WITH HIGHEST SALES/AMOUNT
top_10_states = df.groupby(['ship-state'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='ship-state',y='Amount',data=top_10_states,palette='viridis')
sns.set(rc={"figure.figsize":(20,5)})
plt.show()
```

*From above graph we can see that the highest sales are from Maharastra, Karnataka, and Uttar Pradesh States.

```python
# TOP 10 STATES WITH HIGHEST SALES/AMOUNT
top_10_cities = df.groupby(['ship-city'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='ship-city',y='Amount',data=top_10_cities,palette='viridis')
sns.set(rc={"figure.figsize":(20,5)})
plt.show()
```

*From above graph we can see that the highest sales are from Bengaluru, Hyderabad, and New Delhi.
